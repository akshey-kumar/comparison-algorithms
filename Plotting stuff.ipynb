```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Behavioural predictor performance

```python
acc_array = []
algorithms = ['PCA', 'tsne', 'cebra_hybrid', 'autoencoder', 'autoregressor', 'AbCNet']
for alg in algorithms:
    acc_array.append( np.loadtxt('Saved evaluation metrics/acc_list_'+alg) )
acc_array = np.array(acc_array)
### Neuronal level accuracy (X)
X_acc = np.loadtxt('Saved evaluation metrics/acc_list_X').max()
```

```python
plt.figure(figsize=(5,4))
ax = sns.violinplot(data=acc_array.T[:,:7])
ax.set_xticklabels(['PCA', 'tsne', 'CEBRA-h', 'autoencoder', 'ArAe', 'BunDLe Net'], rotation=40)
ax.set(ylabel = "Decoding accuracy", title ='Behavioural performance of macrovariable')
ax.axhline(0.26, linestyle='--',label='Chance accuracy') ### Chance accuracy (see Evaluation .ipynb)
ax.axhline(X_acc, linestyle='-',label='Neuronal level accuracy', color='red') ### Chance accuracy (see Evaluation .ipynb)

ax.legend()
plt.savefig('behavioural_performance.pdf', transparent=True)
```

```python
plt.figure(figsize=(4.5, 4))
ax = sns.violinplot(data=acc_array.T[:6], linewidth=.5)
ax.set_xticklabels(['PCA', 'tsne', 'CEBRA-h', 'autoencoder', 'ArAe', 'BunDLe-Net'], rotation=40, fontsize=10)
ax.yaxis.set_label_text("Behaviour accuracy", fontsize=11)
ax.axhline(0.26, linestyle='--', label='Chance accuracy', color='gray')
ax.axhline(X_acc, linestyle='-',label='Neuronal level accuracy', color='k') ### Chance accuracy (see Evaluation .ipynb)
plt.grid(axis='y', linestyle=':', color='gray', alpha=0.5)
ax.set_ylim(0, 1)
plt.subplots_adjust(bottom=0.25, left=0.18)  # You can modify this value as needed
ax.legend()
plt.savefig('Figures/behavioural_performance.pdf', transparent=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set_style("whitegrid")

# Customize the color palette
color_palette = sns.color_palette("Set3")

# Create the violin plot
plt.figure(figsize=(4.5, 4))
ax = sns.violinplot(data=acc_array.T[:6], linewidth=1, palette=color_palette)

# Set x-axis tick labels and rotation
ax.set_xticklabels(['PCA', 't-SNE', 'CEBRA-h', 'Autoencoder', 'ArAe', 'BunDLe-Net'], rotation=40, fontsize=10)

# Set y-axis label and styling
ax.set_ylabel("Behavioural Accuracy", fontsize=11)

# Add horizontal lines for chance accuracy and neuronal level accuracy
ax.axhline(0.26, linestyle='--', label='Chance Accuracy', color='gray')
ax.axhline(X_acc, linestyle='-', label='Neuronal Level Accuracy', color='black')

# Customize grid lines
ax.grid(axis='y', linestyle=':', color='gray', alpha=0.5)

# Adjust subplot parameters
plt.subplots_adjust(bottom=0.25, left=0.18)

# Add legend
ax.legend(fontsize=9, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=2)

# Save the figure
plt.savefig('Figures/behavioural_performance.pdf', bbox_inches='tight', transparent=True)

# Show the plot
plt.show()
```

### Dynamics predictor performance

```python
mse_array = []
for alg in algorithms:
    mse_array.append( np.loadtxt('Saved evaluation metrics/mse_list_'+alg) )
mse_array = np.array(mse_array)
goodness_of_fit = 1 -  mse_array[:,:,3]/mse_array[:,:,2]
```

```python
plt.figure(figsize=(4.5, 4))
ax = sns.violinplot(data=goodness_of_fit.T, linewidth=.5)
ax.set_xticklabels(['PCA', 't-SNE', 'CEBRA-h', 'autoencoder', 'ArAe', 'BunDLe-Net'], rotation=40, fontsize=10)
ax.yaxis.set_label_text("Dynamics predictability", fontsize=11)
ax.axhline(0.0, linestyle='--',label='Baseline model', color='gray') ### Baseline goodness of fit (by definition)
plt.grid(axis='y', linestyle=':', color='gray', alpha=0.5)
# Adjust the bottom margin
plt.subplots_adjust(bottom=0.25, left=0.18)  # You can modify this value as needed
ax.legend()
plt.savefig('Figures/dynamical_performance.pdf', transparent=True)
```

```python
sns.set_style("whitegrid")

# Customize the color palette
color_palette = sns.color_palette("Set3")

# Create the violin plot
plt.figure(figsize=(4.5, 4))
ax = sns.violinplot(data=goodness_of_fit.T[:6], linewidth=1, palette=color_palette)

# Set x-axis tick labels and rotation
ax.set_xticklabels(['PCA', 't-SNE', 'CEBRA-h', 'Autoencoder', 'ArAe', 'BunDLe-Net'], rotation=40, fontsize=10)

# Set y-axis label and styling
ax.yaxis.set_label_text("Dynamics predictability", fontsize=11)

ax.axhline(0.0, linestyle='--',label='Baseline model', color='gray') ### Baseline goodness of fit (by definition)

# Customize grid lines
ax.grid(axis='y', linestyle=':', color='gray', alpha=0.5)

# Adjust subplot parameters
plt.subplots_adjust(bottom=0.25, left=0.18)

# Add legend
ax.legend(fontsize=9, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=2)

# Save the figure
plt.savefig('Figures/dynamical_performance.pdf', bbox_inches='tight', transparent=True)

# Show the plot
plt.show()
```

### Comparision of AbC Net with a purely behavioural and purely dynamical macrovaraible

```python
acc_array = []
algorithms = ['supB', 'autoregressor', 'AbCNet']
for alg in algorithms:
    acc_array.append( np.loadtxt('Saved evaluation metrics/acc_list_'+alg) )
acc_array = np.array(acc_array)

mse_array = []
for alg in algorithms:
    mse_array.append( np.loadtxt('Saved evaluation metrics/mse_list_'+alg) )
mse_array = np.array(mse_array)
goodness_of_fit = 1 -  mse_array[:,:,3]/mse_array[:,:,2]
```

```python
ax = sns.barplot(data=[acc_array[0], acc_array[1], acc_array[2]])
ax.set_xticklabels(algorithms)
ax.set(ylabel = "Accuracy", title ='Behavioural performance of macrovariable')
```

```python
ax = sns.barplot(data=[goodness_of_fit[0], goodness_of_fit[1], goodness_of_fit[2]])
ax.set_xticklabels(algorithms)
ax.set(ylabel = "Goodness of fit", title ='Dynamical performance of macrovariable')
```

```python
df = pd.DataFrame([ ['supB', acc_array[0].mean(), goodness_of_fit[0].mean()], 
                    ['autoregressor', acc_array[1].mean(), goodness_of_fit[1].mean()], 
                    ['AbCNet', acc_array[2].mean(), goodness_of_fit[2].mean()]],
                 columns=['algorithm', 'behavioural', 'dynamical'])
                   
df.plot(x='algorithm',
        kind='bar',
        stacked=False)        
                   
                   
```
